# Exploratory Data Analysis of Movies

## Features

* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('movies_complete.csv')
df.set_index('id',inplace=True)
df
```

```python
df['genres'] = df.genres.astype(str)
```

```python
from IPython.display import HTML
```

```python
df_best = df[['poster_path','title','budget_musd','revenue_musd','vote_count','vote_average','popularity']]
```

```python
df_best['Profit'] = df.revenue_musd.sub(df.budget_musd)
df_best['Return'] = df.revenue_musd.div(df.budget_musd)
```

```python
df_best
```

## Renaming Columns for visualization purpose

```python
df_best.columns = ["",'Title','Budget','Revenue','Votes','Average Ratings','Popularity','Profit','ROI']
```

```python
df_best
```

```python
df_best.set_index('Title',inplace=True)
```

```python
df_best
```

### Visualizing posters of movies 

```python
subset = df_best.iloc[:5,:2]
HTML(subset.to_html(escape=False))
```

```python
df_best.Votes.fillna(0,inplace=True)
```

```python
# Function to get the desired movies titles based on its features
def best_worst(n,by,ascending=False,min_budget=0,min_votes=0):
    df_func = df_best.loc[(df_best.Budget>min_budget)&(df_best.Votes>min_votes),['',by]].sort_values(by=by,ascending=ascending).head(n).copy()
    return HTML(df_func.to_html(escape=False))
```

## Top `5` highest rated  movies based on votes greater than `1000`

```python
best_worst(n=5,min_votes=1000,by='Average Ratings')
```

#### Top `5` highest grossing movies based on minimum budget of `$10` `million`

```python
best_worst(n=5,min_budget=10,by="Revenue")
```

#### Top `5` highest budget movies

```python
best_worst(n=5,by="Budget")
```

#### Least `5` movies based on ROI

```python
best_worst(n=5,ascending=True,by='ROI',min_budget=20)
```

#### Top `5` most voted movies

```python
best_worst(n=5,by='Votes',min_votes=2000)
```

# Actors

```python
df.cast
```

```python
act = df.cast.str.split('|',expand=True) #split the strings into lists and in seperate columns
act = act.stack().reset_index(level=1,drop=True).to_frame() # re-organize vertically
act.columns = ['Actor'] # assign column name
act
```

```python
# merging the required columns in the newly created dataframe
act = act.merge(df[['title', 'revenue_musd', 'vote_average', 'popularity']],
                how='left', left_index=True, right_index=True
                )
act
```

```python
act.Actor.nunique()
```

```python
act.Actor.unique()
```

# Most Active Actor

```python
plt.figure(figsize=(15,8))
act.Actor.value_counts().head(10).plot(kind='bar',fontsize=20)
plt.title("Most Active Actor",fontsize=20)
#plt.grid()
plt.yticks(np.arange(0,250,20))
plt.ylabel('No of Movies',fontsize=20);
```

```python
agg_actors = act.groupby('Actor').agg(Total_Revenue=('revenue_musd', 'sum'),
                                      Mean_Revenue=('revenue_musd', 'mean'),
                                      Mean_Rating=('vote_average', 'mean'),
                                      Mean_Pop=('popularity', 'mean'),
                                      Total_Movies=('Actor', 'count'),
                                      )
```

## Most Profitable Actor

```python
top_ten_profitable_actors = agg_actors.nlargest(10,'Total_Revenue').reset_index(level=0)
```

```python
top_ten_profitable_actors
```

```python
most_profitable = agg_actors.Total_Revenue.sort_values(ascending=False)
y_value = top_ten_profitable_actors.Total_Revenue.nlargest(10)
plt.figure(figsize=(15,7))
sns.barplot(x=top_ten_profitable_actors.Actor,y=y_value)
plt.yticks(fontsize=20)
plt.title("Most Profitable Actors",fontsize=20)
plt.ylabel('Total Revenue in millions',fontsize=20)
plt.xlabel('Actors',fontsize=20)
#plt.grid()
plt.xticks(rotation=90,fontsize=20);
```

## Top 10 countries releasing movies

```python
df.production_countries.dropna(inplace=True)
```

```python
country = df.production_countries.str.split('|',expand=True)
country = country.stack().reset_index(level=1,drop=True)
country = country.to_frame()
country.columns = ['Countries']
country
```

```python
countries = country.merge(df[
    'title'], how='left', left_index=True, right_index=True)
countries = countries.groupby('Countries').agg(
    Total_movies=('Countries', 'count')
)
countries = countries.reset_index(level=0)
countries = countries.sort_values(by='Total_movies', ascending=False)
countries['Countries'] = countries['Countries'].replace(
    ['United States of America'], 'USA')
```

```python
x_data = countries.Countries.head(10)
y_data = countries.Total_movies.head(10)
plt.figure(figsize=(15,7))
sns.barplot(x=x_data,y=y_data)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=15)
plt.xlabel('Countries',fontsize=20)
plt.ylabel('Total No of Movies',fontsize=20)
plt.title('Top 10 Countries Releasing Movies',fontsize=25);
```
